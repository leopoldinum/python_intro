## What is Python 

Python is a free and open-source general programming language that was created in 1990 by Guido van Rossum. It is natively object-oriented, though strict OOP is practiced mostly in software engineering. Python is popular in fields related to data science and analytics thanks to some of its useful packages (e.g., Numpy, Pandas).

## Python facts

Python is...

* <b>interpreted</b>, so it does not need to be compiled into machine language
* <b>interactive</b>, allowing programs to be executed and debugged piece by piece
* <b>compatible</b> with many platforms and languages
* <b>easy</b> to read and maintain
* <b>high-level</b>, abstracted from computer processes
* <b>minimalist</b> in style, punctuation, and syntax
* <b>more popular than French</b> in primary schools
* <b>strict about indentation</b>, indifferent about brackets and semicolons
* <b>fast</b>...

```python
# import image packages
from PIL import Image, ImageOps

# function to add border to image
def add_border(input_image, output_image, border):
    img = Image.open(input_image)
    
    if isinstance(border, int) or isinstance(border, tuple):
        bimg = ImageOps.expand(img, border=border)
    else:
        raise RuntimeError('Border is not an integer or tuple!')
    
    bimg.save(output_image)

# call function to create & save image
if __name__ == '__main__':
    in_img = 'C:/Users/natha/OneDrive/Desktop/Python/SP500.jpg'
    add_border(in_img, 
               output_image='C:/Users/natha/OneDrive/Desktop/Python/SP500_border.jpg',
               border=(10,10))
```

### Example - Numpy, Pandas, Matplotlib

```python
# import packages: numpy and pandas
import numpy as np
import pandas as pd

# import matplotlib for plot
from pylab import plt, mpl

# config plot
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline

# read in data; source = https://github.com/yhilpisch/py4fi/blob/master/jupyter36/source/tr_eikon_eod_data.csv
data = pd.read_csv('C:/Users/natha/OneDrive/Desktop/Python/SP500_values.csv',
    index_col=0, parse_dates=True)
data = pd.DataFrame(data['.SPX'])
data.dropna(inplace=True)
data.info()

# calculate log returns, annualized rolling volatility
data['rets'] = np.log(data / data.shift(1))
data['vola'] = data['rets'].rolling(252).std() * np.sqrt(252) 
```

```python
# plot data - uses mpl default axes
data[['.SPX', 'vola']].plot(subplots=True, figsize=(10, 6))
```

### Example - Math, Numpy, NumExpr

mathematical calculations on arrays

```python
import math
loops = 2000000
a = range(1, loops)

def f(x):
    return 3 * math.log(x) + math.cos(x) ** 2

%timeit r = [f(x) for x in a]
```

numpy, use optimized functions

```python
import numpy as np
a = np.arange(1, loops)
%timeit r = 3 * np.log(a) + np.cos(a) ** 2
```

numerical expressions package, compile, avoid in-memory copies of array

```python
import numexpr as ne
 ne.set_num_threads(1)
 f = '3 * log(a) + cos(a) ** 2'
 %timeit r = ne.evaluate(f)
 50.6 ms  4.2 ms per loop (mean  std. dev. of 7 runs, 10 loops each)
```
