## SQL

You can query data in Python using SQL. Many versions of SQL are compatible with Python, but I recommend SQLite because it is serverless and self-contained. This means you do not need to install/run a special SQLite server to use a SQLite database.

```python
import pandas as pd
import sqlite3

con = sqlite3.connect("database.db")
```

```python
df = pd.read_sql_query("SELECT * FROM purchases", con)
df
```

```python
df = df.set_index('index')
df
```

```python
df.to_csv('new_purchases.csv')
```

## Local Files

We can use Python to add a border to the local image shown below.



```python
# import image packages
from PIL import Image, ImageOps

# function to add border to image
def add_border(input_image, output_image, border):
    img = Image.open(input_image)
    
    if isinstance(border, int) or isinstance(border, tuple):
        bimg = ImageOps.expand(img, border=border)
    else:
        raise RuntimeError('Border is not an integer or tuple!')
    
    bimg.save(output_image)

# call function to create & save image
if __name__ == '__main__':
    in_img = 'C:/Users/natha/!Python_intro/SP500.jpg'
    add_border(in_img, 
               output_image='C:/Users/natha/!Python_intro/SP500_border.jpg',
               border=(10,10))

# show image
from IPython.display import Image
Image(filename='C:/Users/natha/!Python_intro/SP500_border.jpg') 
```
